# Area_NEW

```python
doingbiz = pd.read_csv("doingbusiness.csv", encoding="cp1252")
doingbiz["Year"] = doingbiz.Year.str.replace("DB","")
doingbiz["Year"] = doingbiz.Year.astype('int')
```

```python
doingbiz.head()
```

```python
area_data = doingbiz[['Economy','Year','Overall DTF']]
```

```python
area_data
```

```python
def blah(country, measure, year):
    if len(area_data[(area_data.Economy == country) & (area_data.Year == year)][measure].tolist()) == 0: 
        return 0
    else: 
        val = area_data[(area_data.Economy == country) & (area_data.Year == year)][measure].tolist()[0]
        if pd.isnull(val):
            return 0
        else:
            return val
```

```python
columns = ['Economy','Measure_Type','2017','2016','2015','2014','2013','2012','2011','2010','2009','2008','2007','2006','2005','2004']
new_area = pd.DataFrame(columns = columns)
country_list = area_data.Economy.unique()
measures = ['Overall DTF']

#country_list = ["Afghanistan"]
for country in country_list:
    for measure in measures:
        
        df = pd.DataFrame([[country,measure, blah(country, measure, 2017), blah(country, measure, 2016), 
                            blah(country, measure, 2015),
                            blah(country, measure, 2014), blah(country, measure, 2013), blah(country, measure, 2012),
                            blah(country, measure, 2011), blah(country, measure, 2010), blah(country, measure, 2009),
                            blah(country, measure, 2008), blah(country, measure, 2007), blah(country, measure, 2006),
                            blah(country, measure, 2005), blah(country, measure, 2004)]], columns = columns)
                            
        new_area = new_area.append(df, ignore_index=True)
```

```python
new_area.head()
```

```python
new_area.to_csv("slope_doingbiz.csv")
```
